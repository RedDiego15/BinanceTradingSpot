```python
import pandas as pd
```

```python
def createIndex(x):
    key = str(x['Date(UTC)'].year),str(x['Date(UTC)'].day),x['Fee Coin'],str(x['Amount'])
    return key
```

```python
#Amount = cantidad crypto que obtuve
#total = dinero invertido
#Sell_Porcent = porcent of position selled
#price = precio al que compre
#rest valua es el amount de crypto que queda luego de haber vendido un determinado porcentaje de esa posicion 

#df DEBE SER EL OUTPUT 
#df = pd.read_csv("../BinanceDataSet/Exportar historial de operaciones-2022-06-07.csv")
df = pd.read_csv("../Output/CLEAN_data_9-07-2022.csv")
df['Date(UTC)'] = pd.to_datetime(df['Date(UTC)']) 
df['key'] = df.apply(createIndex,axis=1)
df['rest_value'] = df['Amount'] - df['Amount'] * (df['sell_Porcent']/100) 


# #EL RDF DEBE SER DESDE LA FECHA DEL ULTIMO REPORTE HASTA EL DIA anterior 

rdf = pd.read_csv("../BinanceDataSet/Exportar historial de operaciones-2022-09-17 17_58_21.csv")

rdf['Date(UTC)'] = pd.to_datetime(rdf['Date(UTC)'])
rdf['sell_Porcent'] = 0
rdf['is_Open'] = True
rdf['key'] = rdf.apply(createIndex,axis=1)
rdf['rest_value'] = df['Amount'] - df['Amount'] * (df['sell_Porcent']/100) 
df


```

```python
val = df['rest_value'] = df['Amount'] * df['sell_Porcent']
```

```python
rdf
```

```python
df = pd.concat([df,rdf])
df = df.reset_index(drop=True)
df.drop_duplicates(subset=['key'],keep='first',inplace =True)
df = df.loc[df['is_Open']==True]
df
```

```python
dfg = df.groupby('Fee Coin').sum()
dfg
#df['Fee Coin'] == 'BNB'
```

```python
adaDf = df.loc[(df['Market']=='ADAUSDT') | (df['Market']=='ADABUSD')]
btcDf = df.loc[(df['Market']=='BTCUSDT') | (df['Market']=='BTCBUSD')]
ethDf = df.loc[(df['Market']=='ETHUSDT') | (df['Market']=='ETHBUSD')]
dotDf = df.loc[(df['Market']=='DOTUSDT') | (df['Market']=='DOTBUSD')]
solDf = df.loc[(df['Market']=='SOLUSDT') | (df['Market']=='SOLBUSD')]
roseDf = df.loc[(df['Market']=='ROSEUSDT') | (df['Market']=='ROSEBUSD')]



adaDf
```

```python
adaDf = adaDf.loc[adaDf['Type']== 'BUY']
btcDf = btcDf.loc[btcDf['Type']== 'BUY']
ethDf = ethDf.loc[ethDf['Type']== 'BUY']
dotDf = dotDf.loc[dotDf['Type']== 'BUY']
solDf = solDf.loc[solDf['Type']== 'BUY']
roseDf = roseDf.loc[roseDf['Type']== 'BUY']
```

```python
#Order de mayor a menor 
adaDf = adaDf.sort_values(by='Price')
ethDf = ethDf.sort_values(by='Price')
btcDf = btcDf.sort_values(by='Price')
dotDf = dotDf.sort_values(by='Price')
solDf = solDf.sort_values(by='Price')
roseDf = roseDf.sort_values(by='Price')


```

```python
#New Index
btcDf = btcDf.reset_index(drop=True)
dotDf = dotDf.reset_index(drop=True)
adaDf = adaDf.reset_index(drop=True)
ethDf = ethDf.reset_index(drop=True)
solDf = solDf.reset_index(drop=True)
roseDf = roseDf.reset_index(drop=True)

btcDf
```

```python
df.to_csv('../Output/CLEAN_data_9-17-2022.csv',index=False)
adaDf.to_csv('../Output/ADA_clean_data_9-17-2022.csv')
btcDf.to_csv('../Output/BTC_clean_data_9-17-2022.csv')
ethDf.to_csv('../Output/ETH_clean_data_9-17-2022.csv')
dotDf.to_csv('../Output/Dot_clean_data_9-17-2022.csv')
solDf.to_csv('../Output/Sol_clean_data_9-17-2022.csv')
roseDf.to_csv('../Output/Rose_clean_data_9-17-2022.csv')
```
