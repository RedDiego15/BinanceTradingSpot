```python
import pandas as pd
```

```python
def createIndex(x):
    key = str(x['Date(UTC)'].year),str(x['Date(UTC)'].day),x['Fee Coin'],str(x['Amount'])
    return key
```

```python
#Amount = cantidad crypto que obtuve
#total = dinero invertido
#Sell_Porcent = porcent of position selled
#price = precio al que compre
#rest valua es el amount de crypto que queda luego de haber vendido un determinado porcentaje de esa posicion 

#df DEBE SER EL OUTPUT 
#df = pd.read_csv("../BinanceDataSet/Exportar historial de operaciones-2022-06-07.csv")
df = pd.read_csv("../Output/CLEAN_data_01-20-2023.csv")
df['Date(UTC)'] = pd.to_datetime(df['Date(UTC)']) 
df['key'] = df.apply(createIndex,axis=1)
df['rest_value'] = df['Amount'] - df['Amount'] * (df['sell_Porcent']/100) 


# #EL RDF DEBE SER DESDE LA FECHA DEL ULTIMO REPORTE HASTA EL DIA anterior 

rdf = pd.read_csv("../BinanceDataSet/Exportar historial de operaciones-2023-01-25 05_19_41.csv")

rdf['Date(UTC)'] = pd.to_datetime(rdf['Date(UTC)'])
rdf['sell_Porcent'] = 0
rdf['is_Open'] = True
rdf['key'] = rdf.apply(createIndex,axis=1)
rdf['rest_value'] = rdf['Amount'] - rdf['Amount'] * (rdf['sell_Porcent']/100) 
df.head()


```

```python
df = pd.concat([df,rdf])
df = df.reset_index(drop=True)
df.drop_duplicates(subset=['key'],keep='first',inplace =True)
df = df.loc[df['is_Open']==True]
df
```
